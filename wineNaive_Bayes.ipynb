```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.model_selection import train_test_split
from sklearn.utils import shuffle
from sklearn.metrics import accuracy_score
from sklearn.decomposition import PCA
from sklearn.naive_bayes import GaussianNB
#from sklearn.model_selection import cross_val_score
import warnings
warnings.filterwarnings("ignore")
```

```python
data=pd.read_csv("C:\Users\hello\Desktop\WineData\winequality-white.csv",delimiter = ";")
print(data)
#X = np.array(data.loc['fixed acidity':'alcohol'])
#Y = np.array(data.loc['quality'])
quality=data['quality'].values
label1=[]
for i in quality:
    if i<=5:
        label1.append("low") 
    if i==6 or i==7:
        label1.append("medium")
    if i>7:
        label1.append("high")
#print(label1)
d={'label': label1}
df = pd.DataFrame(d)
new_data=data.drop(['quality'], axis=1)
temp=[new_data,df]
final_data=pd.concat(temp,axis=1)
print(final_data)
```

```python
#spliting the data into target and other columns as varibles
X=np.array(final_data.loc[:,'fixed acidity':'alcohol'])
y=np.array(final_data['label'])
X_train, x_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

```python
clf=GaussianNB()
clf.fit(X_train,y_train)
predicted=clf.predict(x_test)
scores=[]
scores.append(accuracy_score(y_test,predicted))
print(scores)
```

```python
#using confusion matrix to find the accuracy
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, predicted)
print(cm)
```

```python
#from statistics import mean 
import time
from sklearn.model_selection import cross_validate
from sklearn.model_selection import ShuffleSplit
scores_test=[]
start=time.time()
for j in range(1,200):
    X=np.array(final_data.loc[:,'fixed acidity':'alcohol'])
    Y=np.array(final_data['label'])
    clf=GaussianNB()
    cv = ShuffleSplit(n_splits=25, test_size=0.2)
    accuracies = cross_validate(clf, X,Y, cv = cv)
    print(accuracies['test_score'])
    print(max(accuracies['test_score']))
    scores_test.append(max(accuracies['test_score']))
end=time.time()
elapsed=end-start
print(scores_test)
mean=sum(scores_test) / len(scores_test)
print("Maximum accuracy after shuffling the data "+str(max(scores_test)))
print("Mean accuracy is",mean)
print("Time taken is :",elapsed)
```
