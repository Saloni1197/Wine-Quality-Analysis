```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.model_selection import train_test_split
from sklearn.utils import shuffle
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import cross_val_score
import warnings
warnings.filterwarnings("ignore")                   
```

```python
data=pd.read_csv("C:\Users\hello\Desktop\WineData\winequality-white.csv",delimiter = ";")
print(data)
X = np.array(data.loc['fixed acidity':'alcohol'])
Y = np.array(data['quality'])
quality=data['quality'].values
label1=[]
for i in quality:
    if i<=5:
        label1.append("low")
    if i==6 or i==7:
        label1.append("medium")
    if i>7:
        label1.append("high")
print(label1)
d={'label': label1}
df = pd.DataFrame(d)
new_data=data.drop(['quality'], axis=1)
temp=[new_data,df]
final_data=pd.concat(temp,axis=1)
print(final_data)
```

```python
features=new_data.columns
print(features)
X=np.array(final_data.loc[:,'fixed acidity':'alcohol'])
Y=np.array(final_data['label'])
```

```python
#finding accuracy using test-train split method
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2)
neighbours=range(1,200)
scores=[]
for i in neighbours:
    model = KNeighborsClassifier(n_neighbors=i)
    model.fit(X_train,y_train)
    predicted= model.predict(X_test)
    scores.append(accuracy_score(y_test,predicted))
print(scores)
maxacc=max(scores)
print(maxacc)
print("Maximum accuracy when taking 20 percent as test data"+str(maxacc))
k=neighbours[scores.index(maxacc)]
print("K optimal values is :"+str(k))
```

```python
#finding accuracy using cross validate method
import time
from sklearn.model_selection import cross_validate
from sklearn.model_selection import ShuffleSplit
import numpy as np
neighbours=range(1,100)
scores=[]
#create a new KNN model
start=time.time()
for i in neighbours:
    model = KNeighborsClassifier(n_neighbors=i)
    cv = ShuffleSplit(n_splits=25, test_size=0.2)
    accuracies = cross_validate(model, X,Y, cv = cv)
    print(accuracies['test_score'])
    print(max(accuracies['test_score']))
    scores.append(max(accuracies['test_score']))
end=time.time()
k=neighbours[scores.index(max((scores)))]
elapsed=end-start
print("Maximum accuracy while taking "+str(k)+"th neighbour is :"+str(max([scores])))
print("K optimal values is :"+str(k))
mean=sum(scores) / len(scores)
print("Mean accuracy is :",mean)
print("Time taken to train the model is:",elapsed)
```
